```python
myInt = int(10)
myFloat = float(20.5)
myBool = bool(True)
myStr = str("Biba")
myList = [1, 2, 3, 4, 5]

print(f"myint:{myInt} {type(myInt)}")
print(f"myFloat: {myFloat} {type(myFloat)}")
print(f"myBool: {myBool} {type(myBool)}")
print(f"myStr: {myStr} {type(myStr)}")
print(f"myList: {myList} {type(myList)}")
myint:10 <class 'int'>
myFloat: 20.5 <class 'float'>
myBool: True <class 'bool'>
myStr: Biba <class 'str'>
myList: [1, 2, 3, 4, 5] <class 'list'>
def addition():
    a = float(input("Введіть перше число: "))
    b = float(input("Введіть друге число: "))
    c = a + b
    print("Результат додавання", c)


def minus():
    a = float(input("Введіть перше число: "))
    b = float(input("Введіть друге число: "))
    c = a - b
    print("Результат віднімання", c)


def multiplication():
    a = float(input("Введіть перше число: "))
    b = float(input("Введіть друге число: "))
    c = a * b
    print("Результат множення: ", c)


def division():
    a = float(input("Введіть перше число: "))
    b = float(input("Введіть друге число: "))

    if b != 0:
        c = a / b
        print("Результат ділення:", c)
    else:
        print("На нуль не ділиться")


def choose_case():
    while True:
        try:
            choose = int(input(
            "Яку дію хочете виконати?\n 1. Додавання\n 2. Віднімання\n 3. Множення\n 4. Ділення \n 5. Вихід \n Обирайте:"))
            if choose == 1:
                addition()
            elif choose == 2:
                minus()
            elif choose == 3:
                multiplication()
            elif choose == 4:
                division()
            elif choose == 5:
                break
            else:
                print("Error")
        except ValueError:
            print("помилка вводу")

choose_case()
def addition():
    a = float(input("Введіть число до якого потрібно додати: "))
    b = float(input("Введіть число, яке потрібно додати: "))
    c = a + b
    print("Результат додавання", c)


def multiplication():
    a = float(input("Введіть число, яке потрібно помножити: "))
    b = float(input("Введіть число на яке потрібно помножити: "))
    c = a * b
    print("Результат множення: ", c)


def division():
    a = float(input("Введіть перше число: "))
    b = float(input("Введіть друге число: "))

    if b != 0:
        c = a / b
        print("Результат ділення:", c)
    else:
        print("На нуль не ділиться")


def comparison():  
    a = float(input("Введіть число, яке потрібно порівнювати: "))
    b = float(input("Введіть число, з яким потрібно порівнювати "))
    if a > b:
        print("Число ", a, "більше")
    elif a == b:
        print("Числа дорівнюють одне одному")
    else:
        print("Число ", b, "більше")


def power():
    a = float(input("Введіть число, яке хочете піднести до степення: "))
    b = float(input("До якого числа потрібно піднести даний степінь? "))
    c = a ** b
    print("Результат: ", c)


def remainder():
    a = float(input("Введіть число, яке потрібно поділити: "))
    b = float(input("Введіть число на яке потрібно поділити: "))

    if b != 0:
        c = a % b
        print("Остача від ділення дорівнює: ", c)
    else:
        print("Ділення на нуль неможливе")


def choose_case():
    while True:
        try:
            choose = int(input(
            "Яку дію хочете виконати?\n 1. Додавання \n 2. Віднімання \n 3. Множення \n 4. Ділення \n 5. Порівняння\n 6. Піднесення до степення\n 7. Остача від ділення\n 8. Вихід\n Обирайте:"))
            if choose == 1:
                addition()
            elif choose == 2:
                minus()
            elif choose == 3:
                multiplication()
            elif choose == 4:
                division()
            elif choose == 5:
                comparison()
            elif choose == 6:
                power()
            elif choose == 7:
                remainder()
            elif choose == 8:
                print("Вихід")
                break
            else:
                print("Error")
        except ValueError:
            print("Помилка вводу")

choose_case()
```
